```python
import warnings 
warnings.simplefilter('ignore')

import scanpy as sc
import scparadise
import pandas as pd
import os
```

```python
# Load normalized integrated data
adata = sc.read_h5ad('Mouse_aging_brain/adata_unintegrated.h5ad')
```

```python
# Create lists with paired samples and annotation levels
lst_test = ['young4', 'old2', 'old4', 'young1', 'oldex4']
```

```python
# Cell type prediction evaluation
for folder in lst_test:
    # Create adata_test
    adata_test = adata[adata.obs['orig.ident'].isin([folder])].copy()
    # Load metadata predicted using Azimuth 
    meta = pd.read_csv(os.path.join('Mouse_aging_brain/Azimuth/reports', folder, 'meta.csv'), index_col=0)
    # Add metadata to adata_test object
    adata_test.obs = meta
    # Create and save classification report of annotation levels
    scparadise.scnoah.report_classif_full(adata_test, 
                                          celltype = 'Celltype', 
                                          pred_celltype='predicted.Celltype', 
                                          report_name = 'report_test_Azimuth_3000_celltype_l1.csv',
                                          save_path = os.path.join('Mouse_aging_brain/Azimuth/reports', folder),
                                          save_report=True)
```
